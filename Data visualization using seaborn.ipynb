```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("nba.csv")
```

```python
data.sample(5)
```

# Line Plot

```python
sns.lineplot(data['Age'], data['Weight'])
```

```python
sns.lineplot(x=data['Age'], y=data['Weight'])
```

```python
sns.lineplot(x=data["Age"] , y= data["Weight"] , hue = data["Position"] )
```

# Scatter Plot

```python
sns.scatterplot(x= data["Age"] , y= data["Weight"])
```

```python
sns.scatterplot(x= data["Age"] , y= data["Weight"] , hue = data["Position"])
```

# Box Plot

```python
sns.boxplot(x=data["Age"])
```

```python
sns.boxplot(x=data["Age"] , y=data["Weight"])
```

# Violin Plot

```python
sns.violinplot(x=data["Age"])
```

```python
sns.set(style = "whitegrid")
sns.violinplot(x=data["Age"] , y=data["Weight"])
```

# Swarm Plot

```python
sns.set(style = "whitegrid")
sns.swarmplot(x=data["Age"])
```

```python
sns.set(style = "whitegrid")
sns.swarmplot(x="Age", y= "Weight" , data= data)
```

# Bar Plot

```python
sns.set(style = "whitegrid")
sns.barplot(x=data["Age"])
```

```python
sns.set(style = "whitegrid")
sns.barplot(x=data["Age"] , y=data["Weight"])
```

# Point Plot

```python
sns.set(style = "whitegrid")
sns.pointplot(x=data["Age"] , y = data["Weight"])
```

# Count Plot

```python
sns.set(style = "whitegrid")
sns.countplot(x=data["Age"])
```

# KDE Plot

```python
sns.kdeplot(x= data["Age"])
```

```python
sns.kdeplot(x= data["Age"] , y= data["Number"])
```

```python
sns.distplot(x=data["Age"])
```
